```python
lsdc=[]
import operator
for k,G in GG.items():  
    d=nx.degree_centrality(G)

    m=max(d.iteritems(), key=operator.itemgetter(1))[0]
    maxcent=d.get(m)
    
    n=nx.nodes(G)

    s=0
    for i in range(len(d)):
        s=s+maxcent-d.values()[i]
        
    lsdc.append(s)
    
print lsdc,len(lsdc)
```

```python

lscc=[]
import operator
for k,G in GG.items():  
    d=nx.closeness_centrality(G)

    m=max(d.iteritems(), key=operator.itemgetter(1))[0]
    maxcent=d.get(m)
    
    n=nx.nodes(G)

    s=0
    for i in range(len(d)):
        s=s+maxcent-d.values()[i]
        
    lscc.append(s)
    
print lscc,len(lscc)
```

```python
lsbc=[]
for k,G in GG.items():  
    d=nx.betweenness_centrality(G)

    m=max(d.iteritems(), key=operator.itemgetter(1))[0]
    maxcent=d.get(m)
    
    n=nx.nodes(G)

    s=0
    for i in range(len(d)):
        s=s+maxcent-d.values()[i]
        
    lsbc.append(s)
    
print lsbc,len(lsbc)

```

```python
import math
lskc=[]
for k,G in GG.items():  
    G=nx.Graph(G)
    phi = (1+math.sqrt(5))/2.0 # largest eigenvalue of adj matrix
    d=nx.katz_centrality_numpy(G,1/phi-0.01)
    
    m=max(d.iteritems(), key=operator.itemgetter(1))[0]
    maxcent=d.get(m)
    
    n=nx.nodes(G)

    s=0
    for i in range(len(d)):
        s=s+maxcent-d.values()[i]
        
    lskc.append(s)
    
print lskc,len(lskc)
```

```python
lsec=[]
for k,G in GG.items():  
    d=nx.eigenvector_centrality(G,max_iter=2000)
    m=max(d.iteritems(), key=operator.itemgetter(1))[0]
    maxcent=d.get(m)
    
    n=nx.nodes(G)

    s=0
    for i in range(len(d)):
        s=s+maxcent-d.values()[i]
        
    lsec.append(s)

print lsec,len(lsec)
```

```python
lspc=[]
for k,G in GG.items():  
    d=nx.pagerank(G,alpha=0.9)

    m=max(d.iteritems(), key=operator.itemgetter(1))[0]
    maxcent=d.get(m)
    
    n=nx.nodes(G)

    s=0
    for i in range(len(d)):
        s=s+maxcent-d.values()[i]
        
    lspc.append(s)

print lspc,len(lspc)
```

```python
nGG=[]
for k,G in GG.items():
    nGG.append(G.name)
print nGG
```

```python
ddd=zip(nGG,lsdc,lsbc,lscc,lskc,lsec,lspc)
print ddd
```

```python
cdf=pd.DataFrame(ddd)
cdf.columns = ['cut hashtag graph', 'degree centralization','betweenness centralization', 'closeness centralization', 'katz centralization','eigenvector centralization','pagerank centralization','cut-off values']
cdf['cut-off values']=GG.keys()
cdf=cdf.sort(['cut-off values'], ascending=[True])
cdf
```
